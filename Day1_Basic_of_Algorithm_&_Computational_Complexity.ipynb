#### 1. 선형 탐색 (Linear Search)
가장 기본이 되는 탐색 알고리즘입니다.


데이터가 무작위로 섞여 있을 때, 우리가 원하는 값을 찾기 위해 처음부터 끝까지 하나씩 확인하는 방식입니다.

💡 **요구사항 및 로직 설명**

- 함수 정의: linear_search(list, value) 형태로 작성합니다.

- 순차 방문: 리스트의 첫 번째 요소부터 마지막 요소까지 인덱스를 활용해 접근합니다. (for문 사용)

- 비교 연산: 현재 확인 중인 요소(list[i])가 찾고자 하는 값(value)과 일치하는지 확인합니다.

- 반환값:

    - 찾았다면? 해당 요소의 **인덱스(i)**를 즉시 반환하고 종료합니다.

    - 끝까지 못 찾았다면? 리스트에 값이 없다는 의미로 **-1**을 반환합니다.

```python
def linear_search(list, value):
    for i in range(len(list)):
        if list[i] == value:
            return i
    return -1
```

📊 **시간 복잡도 분석**

- 데이터의 개수를 $N$이라고 할 때, 운이 나쁘면 끝까지 다 확인해야 하므로 비교 연산(==)은 최대 $N$번 발생합니다.
- 따라서 시간 복잡도는 **$O(N)$**입니다.

#### 2. 선택 정렬 (Selection Sort)

선택 정렬은 전체 데이터 중 가장 작은 값(최솟값)을 '선택'해서 맨 앞으로 보내는 과정을 반복하는 정렬 방식입니다.

💡 **요구사항 및 로직 설명**
- 이중 반복문 사용:

    1. 외부 루프: 현재 최솟값을 채워 넣을 위치(i)를 결정합니다. (0부터 리스트 끝까지)

    2. 내부 루프: 아직 정렬되지 않은 나머지 부분(i+1부터 끝까지)에서 가장 작은 값을 찾습니다.

- 최솟값 추적: smallest라는 변수에 현재까지 찾은 가장 작은 값의 인덱스를 저장하며 비교합니다.

- 값 교환 (Swap): 내부 루프가 끝나면, 결정된 최솟값(list[smallest])과 현재 위치의 값(list[i])을 서로 바꿉니다. 파이썬에서는 a, b = b, a 문법으로 간단히 바꿀 수 있습니다.

📝 **작성 단계**
1. 함수 선언: selection_sort 함수를 만들고 정렬할 리스트(list)를 인자로 받습니다.

2. 전체 범위 루프 (Outer Loop): 리스트의 처음부터 끝까지 한 칸씩 이동하며 '가장 작은 값이 들어올 자리'를 지정합니다.

    - 변수 i를 사용하여 for i in range(len(list))를 작성하세요.

3. 최솟값 가정: 일단 현재 자리(i)에 있는 값이 가장 작다고 가정합니다.

    - 그 위치의 인덱스를 기억할 변수 smallest = i를 만드세요.

4. 탐색 범위 루프 (Inner Loop): 현재 위치(i)의 다음 칸부터 끝까지 뒤져보며 더 작은 값이 있는지 찾습니다.

    - for j in range(i + 1, len(list))를 사용하여 범위를 지정하세요.

5. 최솟값 갱신: 만약 j번째 칸의 값(list[j])이 우리가 알고 있던 최솟값(list[smallest])보다 더 작다면?

    - smallest 변수에 새로운 인덱스 j를 저장합니다.

6. 위치 바꾸기 (Swap): 안쪽 루프가 끝나면 smallest에는 진짜 최솟값의 위치가 담겨 있습니다. 이제 처음 가정했던 자리 i와 실제 최솟값 자리 smallest의 데이터를 서로 바꿉니다.

    - 파이썬의 특징인 a, b = b, a 방식을 사용해 보세요.

```python
def sort_selection(list):
    for i in range(len(list)):
        smallest = i
        for j in range(i+1, len(list)):
            if list[j] < list[smallest]:
                smallest = j
        list[i], list[smallest] = list[smallest], list[i]

    return list
```

📊 **시간 복잡도 분석**

외부 루프가 $N$번 돌 때, 내부 루프는 $N, N-1, N-2, \dots$ 번씩 점차 줄어들며 수행됩니다.

전체 연산 횟수는 약 $\frac{N(N-1)}{2}$번이며, 가장 큰 차수인 $N^2$을 따라 시간 복잡도는 **$O(N^2)$**이 됩니다.

* 나의 접근

리스트를 새로 만들어 그곳에 가장 최솟값들 부터 채워넣기

```python
def selection_sort(put_list):
    result = []
    if put_list:
        for i in range(len(put_list)):
            result.append(min(put_list))
            put_list.remove(min(put_list))
    return result
```

```python
selection_sort([3, 6, 1, 2, 5, 10])
```

* 교환정렬

첫번째 엘리먼트와 그 뒤의 엘리먼트의 값의 크기를 비교한다,

가장 작을때까지,

비교대상이 더 작다면 두 수의 자리를 바꾼다.

```python
list_a = [3, 6, 1, 2, 5, 10]

for i in range(len(list_a)):

    smallest = i

    for j in range(i+1, len(list_a)):

        if list_a[j] < list_a[smallest]:

            smallest = j

    list_a[i], list_a[smallest] = list_a[smallest], list_a[i]

print(list_a)
```

## 점근적 분석 그래프

이 그래프는 데이터($n$)가 늘어날 때 알고리즘의 실행 시간이 얼마나 폭발적으로 증가하는지를 한눈에 보여줍니다.

그래프의 계급도 (그래프 상에서 아래 있을수록 우수한 알고리즘)
1. $\log_2 n$ (Logarithmic): 데이터가 2배 늘어도 시간은 아주 조금 늘어납니다. (예: 이진 탐색)
2. $n$ (Linear): 데이터와 시간이 정비례합니다. (예: 선형 탐색)
3. $n \cdot \log_2 n$ (Log-linear): 효율적인 정렬 알고리즘의 속도입니다. (예: 퀵 정렬, 병합 정렬)
4. $n^2$ (Quadratic): 우리가 아까 만든 선택 정렬의 속도입니다. 데이터가 10배 늘면 시간은 100배 늘어납니다.
5. $n^3$ (Cubic): 3중 반복문이 쓰일 때 발생하며, 매우 느려집니다.
6. $2^n$ (Exponential): **"절대 피해야 할 구간"**입니다. $n$이 조금만 커져도 슈퍼컴퓨터로도 해결이 불가능해집니다.

1. 왜 $\log_2$가 나타날까? (핵심 원리)

알고리즘에서 $\log_2$는 주로 **"반으로 나누는 과정"**이 있을 때 등장합니다.

- 상황: 데이터 $N$개가 있을 때, 한 번 확인할 때마다 탐색 범위가 절반씩 줄어드는 경우입니다.
- 수학적 이유: $N$을 몇 번 반으로 나누어야 1이 되는지를 계산하는 것이 바로 $\log_2 N$입니다.
- 데이터가 8개일 때: $8 \rightarrow 4 \rightarrow 2 \rightarrow 1$ (3단계, 즉 $\log_2 8 = 3$)
- 데이터가 1024개일 때: 단 10번만 확인하면 원하는 값을 찾을 수 있습니다 ($\log_2 1024 = 10$).
- 대표 사례: 이진 탐색(Binary Search). 업다운 게임을 할 때 범위를 절반씩 줄여나가는 것과 같습니다.
---
2. 그래프의 곡선별 발생 상황 (복잡도 계급)
이미지에서 보신 그래프의 곡선들은 다음과 같은 프로그래밍 상황에서 나타납니다.

① $O(1)$ - 상수 시간 (그래프의 가장 바닥)
- 상황: 데이터의 양에 상관없이 딱 한 번에 끝나는 작업입니다.
- 예시: 리스트에서 특정 인덱스의 값을 가져오기 (list[5]), 변수에 값 대입하기.

② $O(\log_2 n)$ - 로그 시간
- 상황: 문제를 해결할 때마다 확인해야 할 데이터가 절반씩 뚝뚝 떨어져 나가는 경우입니다.
- 예시: 정렬된 사전에서 단어 찾기, 이진 탐색 트리 활용.
③ $O(n)$ - 선형 시간
- 상황: 리스트의 처음부터 끝까지 모든 데이터를 한 번씩 다 훑어야 하는 경우입니다.
- 예시: 아까 연습했던 선형 탐색(Linear Search), 리스트의 모든 숫자 합 구하기.

④ $O(n \log_2 n)$ - 선형 로그 시간
- 상황: 효율적인 정렬 알고리즘에서 주로 나타납니다. 데이터를 반으로 나누어 각각 정렬한 뒤 다시 합치는 방식입니다.
- 예시: 퀵 정렬(Quick Sort), 병합 정렬(Merge Sort).

⑤ $O(n^2)$ - 2차 시간
- 상황: 데이터 하나당 전체 데이터를 다시 한번 다 훑어야 하는 중첩 반복문 상황입니다.
- 예시: 학생이 직접 구현한 선택 정렬(Selection Sort), 이중 for문.

⑥ $O(2^n)$ - 지수 시간
- 상황: 한 단계를 진행할 때마다 선택지가 2배씩 늘어나는 경우입니다.
- 예시: 피보나치 수열의 재귀 구현, 모든 가능한 조합을 다 찾는 브루트 포스(Brute-force).
---
3. 요약: 엔지니어가 이 숫자를 보는 법
- $O(1)$ ~ $O(n \log n)$: 실무에서 사용하기에 아주 좋은 알고리즘입니다.
- $O(n^2)$: 데이터가 조금만 많아져도(예: 10만 개 이상) 시스템이 멈출 수 있어 주의해야 합니다.
- $O(2^n)$: 데이터가 30~40개만 되어도 계산에 수만 년이 걸릴 수 있는 위험한 알고리즘입니다.
